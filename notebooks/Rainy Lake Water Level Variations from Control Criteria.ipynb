# Rainy Lake Water Level Variations from Control Criteria

## Import Water Level Data

```python
%matplotlib inline

# Standard Python modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import datetime

RL = pd.read_pickle('../data/RL.pkl')

RL.name = 'RAINY LAKE (RECONCILED)'

fig,ax = plt.subplots(1,1,figsize=(10,5))
RL.plot(lw=1)
plt.ylabel('level [meters]')
plt.title(RL.name \
          + ' from ' + RL.index.min().strftime('%m-%d-%Y') \
          + ' to ' + RL.index.max().strftime('%m-%d-%Y'))

```

## Control Critera for Loons

Water level can rise up to 0.15m and drop up to 0.30m during May through July.

## Control Criteria for Muskrats

From October 15 through November 15, the water level must be less than 1m.
From November 15 through February 28, the water level must not rise more than 0.15m or drop more than 0.60m. 

## Control Criteria for Wild Rice

Between April 20 and May 15, the water level must be between 0.50m and 1.2m.
Between May 15 and July 1, the water levels should not increase or decrease by more than 0.15m.
